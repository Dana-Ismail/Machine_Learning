```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PowerTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from pprint import pprint
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn_pandas import DataFrameMapper
from sklearn import preprocessing
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn2pmml import PMMLPipeline
from sklearn.preprocessing import Normalizer
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv ('data.csv')
```

```python
numeric_df = df[['engine_power','speedometer','previous_owner','price']]
categorical_df = df[['model_year','name','color','fuel','origin','gear_type', 'status','bi_condition','bi_glass','bi_license', 'bi_radio', 'bi_alert','bi_cushion','bi_sunroof','bi_wheels','bi_seat_type','bi_payment_option','bi_lock_system']]
```

```python
pt = PowerTransformer()
pt.fit(numeric_df)
numeric_df_ = pd.DataFrame()
numeric_df_ = pt.transform(numeric_df)
```

```python
numeric_data = pd.DataFrame(numeric_df_,
columns= ['engine_power','speedometer','previous_owner','price'])
```

```python
LE = LabelEncoder()
model_year_column = df['model_year']
model_year_column = LE.fit_transform(model_year_column)
```

```python
column_transformer = OneHotEncoder(handle_unknown='ignore',sparse=False)

transformed_data = column_transformer.fit_transform(categorical_df)
transformed_categorical_df = pd.DataFrame(transformed_data, columns=column_transformer.get_feature_names(['name','color','fuel','origin','gear_type', 'status','bi_condition','bi_glass','bi_license', 'bi_radio', 'bi_alert','bi_cushion','bi_sunroof','bi_wheels','bi_seat_type','bi_payment_option','bi_lock_system']))
transformed_categorical_df
```

```python
modified_df = pd.DataFrame()
modified_df = pd.concat([numeric_data, transformed_categorical_df], axis=1)
```

```python
modified_df.drop(6862, inplace=True)
```

```python
x_train, x_test, y_train, y_test = \
    train_test_split(modified_df.drop(columns=['price']),
                     modified_df['price'],
                     test_size=0.2,
                     random_state=42)
```

```python
gbt = GradientBoostingRegressor(random_state = 42)
pprint(gbt.get_params())
```

```python
gbt_random = RandomizedSearchCV(estimator = gbt,
                               param_distributions = random_grid_search,
                               n_iter = 10, 
                               cv = 5,
                               verbose=2,
                               random_state=42,
                               n_jobs = 1)
gbt_random
```

### Pipeline

```python
data = pd.read_csv ('data.csv')
```

```python
data
```

```python
numeric_features = ['engine_power','speedometer','previous_owner']
categorical_features = ['name','model_year','color','fuel','origin','gear_type','status','bi_condition','bi_glass','bi_license', 'bi_radio', 'bi_alert','bi_cushion','bi_sunroof','bi_wheels','bi_seat_type','bi_payment_option','bi_lock_system']
numeric_preprocessing_steps = Pipeline([
    ('transformer', PowerTransformer())
])

categorical_preprocessing_steps = Pipeline([
    ('onehot', OneHotEncoder())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_preprocessing_steps, numeric_features),
        ('cat', categorical_preprocessing_steps, categorical_features)
    ]
)

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('GradientBoostingRegressor',GradientBoostingRegressor(max_depth=5, min_samples_leaf=7, min_samples_split=16,
                          random_state=42))
])
```

```python
modified_df.drop(['price'],axis=1,inplace=True)
```

```python
y_train=pd.DataFrame(data['price'])
x_train=pd.DataFrame(data.drop(['price'],axis=1))
```

```python
y_train = pd.DataFrame(y_train)
```

```python
pt = PowerTransformer()
pt.fit(y_train)
target_variable = pd.DataFrame()
target_variable = pt.transform(y_train)
```

```python
pipeline.fit(x_train, target_variable)
```
